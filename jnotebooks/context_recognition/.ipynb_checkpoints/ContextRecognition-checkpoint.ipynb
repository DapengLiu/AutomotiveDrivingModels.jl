# Context Recognition

```julia
using AutomotiveDrivingModels
using RandomForestBehaviors
using DynamicBayesianNetworkBehaviors

include(Pkg.dir("AutomotiveDrivingModels", "scripts", "extract_params.jl"))
include(Pkg.dir("AutomotiveDrivingModels", "scripts", "model_params.jl"));
```

```julia
include(Pkg.dir("AutomotiveDrivingModels", "viz", "incl_cairo_utils.jl"))
include(Pkg.dir("AutomotiveDrivingModels", "viz", "incl_pgfplots_utils.jl"))
```

```julia
context_classes = ["freeflow", "following", "lanechange"]
ncontext_classes = length(context_classes)

println("Loading Datasets: "); tic()
dsets = Dict{AbstractString, ModelTrainingData2}() # context_class target -> dset
streetnets = Dict{AbstractString, StreetNetwork}()
cv_splits = Dict{AbstractString, FoldAssignment}() # context class target -> cv_splut
trace_likelihoods = Dict{AbstractString, Matrix{Float64}}() # context class target -> [context_class_used_to_pred × ntraces]
trace_counts = Dict{AbstractString, Matrix{Int}}() # context class target -> [context_class_used_to_pred × ntraces]
for context_class in context_classes
    dataset_jld_file = joinpath(EVALUATION_DIR, "dataset2_" * context_class * ".jld")
    dset = JLD.load(dataset_jld_file, "model_training_data")::ModelTrainingData2

    cv_split = get_fold_assignment_across_drives(dset, N_FOLDS)

    dsets[context_class] = dset
    cv_splits[context_class] = cv_split
    trace_likelihoods[context_class] = fill(NaN, ncontext_classes, length(cv_split.seg_assignment)) # context_class_train × ntraces
    trace_counts[context_class] = zeros(Int, ncontext_classes, length(cv_split.seg_assignment))
end
toc()

nothing
```

```julia
context_class_true = context_classes[3]

println("True Context Class: ", context_class_true)

dset = dsets[context_class_true]
runlogs_test = load_runlogs(dset)
streetnets = load_streetnets(runlogs_test)
n_segs = length(dset.runlog_segments)
```

### Choose Segment and Display It

```julia
segind = 72
seg = dset.runlog_segments[segind]
runlog = runlogs_test[seg.runlog_id]
sn = streetnets[runlog.header.map_name]

plot_manipulable_runlog(runlog, sn, frames=seg.frame_start:seg.frame_end)
```

```julia
include(Pkg.dir("AutomotiveDrivingModels", "viz", "incl_pgfplots_utils.jl"))
# compute likelihood under each model

segind = 1
# model_name = "Static Gaussian"
model_name = "Linear Bayesian"

@manipulate for segind in 1:length(dset.runlog_segments)

    seg = dset.runlog_segments[segind]
    runlog = runlogs_test[seg.runlog_id]
    sn = streetnets[runlog.header.map_name]

    model_short_name = convert_model_name_to_short_name(model_name)

    folds = 1 : N_FOLDS
    fold = folds[1]

    models = Dict{AbstractString, AbstractVehicleBehavior}()

    eval_model_names = AbstractString[]
    eval_trace_logls = Float64[]

    for context_class_model_is_from in context_classes
        dset_filepath_modifier = "_" * context_class_model_is_from
        model_for_fold_path_jld = joinpath(EVALUATION_DIR, model_short_name * dset_filepath_modifier * "_fold" * @sprintf("%02d", fold) * ".jld")
        model = JLD.load(model_for_fold_path_jld, "model")
        eval_model_name = context_class_model_is_from*" "*string(fold)
        push!(eval_model_names, eval_model_name)
        models[eval_model_name] = model
        push!(eval_trace_logls, calc_trace_likelihood(runlog, sn, seg, model))
    end

    highest_likelihood_index = indmax(eval_trace_logls)
    for i in 1 : length(eval_trace_logls)

        if context_classes[i] == context_class_true
            print("→ ")
        else
            print("  ")
        end

        @printf("%12s %10.3f", eval_model_names[i], eval_trace_logls[i])
        if i == highest_likelihood_index
            println(" ←")
        else
            println("")
        end
    end

    ax = plot_logl_vs_trace(eval_model_names, models, runlog, sn, seg.frame_start, seg.frame_end, ID_EGO, include_legend_entry=true)
    push!(ax.plots, Plots.Node(context_classes[highest_likelihood_index],0.1,0.1,style="right",axis="axis description cs"))
    ax.legendPos = "outer north east"
    ax
end
```

```julia
dset_filepath_modifier = "_following"
fold = 5
model_short_name = convert_model_name_to_short_name(model_name)
model_for_fold_path_jld = joinpath(EVALUATION_DIR, model_short_name * dset_filepath_modifier * "_fold" * @sprintf("%02d", fold) * ".jld")
model = JLD.load(model_for_fold_path_jld, "model")
println("nparents: ", length(get_indicators(model)))
for f in get_indicators(model)
    println(symbol(f))
end
```
